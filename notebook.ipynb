```python
import pandas as pd
import math
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_excel('bundestag_full.xlsx')
```

```python
data.Votes = data.Votes.apply(lambda x: x.replace('.', ''))
```

```python
# bt_votes.to_excel(excel_writer = path, index = False, engine = 'xlsxwriter')
```

```python
data.Votes = data.Votes.astype(int)
```

```python
data.columns = data.columns.str.strip()
```

```python
# data['votes_%'] = data
```

```python
votes_seats_sum = data.groupby('Year', as_index = False) \
    .agg({'Votes' : 'sum', 'Seats' : 'sum'}) \
    .rename(columns = {'Votes' : 'votes_sum', 'Seats' : 'seats_sum'})
```

```python
votes_seats_sum
```

```python
bt = data.merge(votes_seats_sum, on = 'Year')
```

```python
bt['votes_'] = round((bt.Votes / bt.votes_sum) * 100, 2)
bt['seats_'] = round((bt.Seats / bt.seats_sum) * 100, 2)
```

```python
bt['md'] = abs(bt.votes_ - bt.seats_)
```

## Коалиции 2017 года

```python
import itertools

party_seats = [246, 153, 67, 80, 69, 94]


def col_count(parties, amount):
    coalitions = []
    a = list(itertools.combinations(parties, amount))
    for party in a:
        b = ''.join(party)
        coalitions.append(b)
    votes = []
    summ = 0
    for coalition in coalitions:
        if coalition.find('A') != -1:
            summ += party_seats[0]
        if coalition.find('B') != -1:
            summ += party_seats[1]
        if coalition.find('C') != -1:
            summ += party_seats[2]
        if coalition.find('D') != -1:
            summ += party_seats[3]
        if coalition.find('E') != -1:
            summ += party_seats[4]
        if coalition.find('F') != -1:
            summ += party_seats[5]
        votes.append(summ)
        summ = 0
    return votes, coalitions


a = col_count('ABCDEF', 2)
b = col_count('ABCDEF', 3)
c = col_count('ABCDEF', 4)
d = col_count('ABCDEF', 5)
e = col_count('ABCDEF', 6)
votes = a[0] + b[0] + c[0] + d[0] + e[0]
coalitions = a[1] + b[1] + c[1] + d[1] + e[1]
table_ = {'coalitions':coalitions, 'votes': votes}
table = pd.DataFrame(table_)
table_quote = table[table.votes > 355]
table_quote
```
